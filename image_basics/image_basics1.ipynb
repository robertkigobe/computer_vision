```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
path = '/Users/robertkigobe/Documents/My_Research/My_Python/Computer_vision/images/norumbega.png'
```

```python
img = cv2.imread(path)
img.shape
```

```python
#opencv reads image as blue green red
plt.imshow(img)
```

```python
#transform picture to RGB
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
#read in a gray image
img_gray = cv2.imread(path,cv2.IMREAD_GRAYSCALE)
img_gray.shape #shows it only has two channels
```

```python
plt.imshow(img_gray,cmap='gray')
```

```python

fix_img.shape
```

```python
#resize image to new dimensions
new_img = cv2.resize(fix_img,(2500,1500))
plt.imshow(new_img)
```

```python
#resize by ratios
w_ratio = 0.5
h_ratio = 0.5
```

```python
ratio_img = cv2.resize(fix_img,(0,0),fix_img,w_ratio,h_ratio)
```

```python
plt.imshow(ratio_img)
```

```python
#flip images vertically
flip_img1 = cv2.flip(fix_img,0)
plt.imshow(flip_img1)
```

```python
#flip images horizontally 
flip_img2 = cv2.flip(fix_img,1)
plt.imshow(flip_img2)
```

```python
#flip images horizontally and vertically
flip_img3 = cv2.flip(fix_img,-1)
plt.imshow(flip_img3)
```

```python
#save image
cv2.imwrite('/Users/robertkigobe/Documents/My_Research/My_Python/Computer_vision/images/new_norumbega.jpg',flip_img3)
```

```python
#display images on the canvas space
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
ax.imshow(flip_img3)
```
