```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#path = '/Users/robertkigobe/Documents/My_Research/My_Python/Computer_vision/images/norumbega.png'
```

```python
blank_img = np.zeros(shape=(512,512,3), dtype=np.int16)
plt.imshow(blank_img)
```

```python
#draw recotrangels
cv2.rectangle(blank_img,pt1=(384,0),pt2=(510,150),color=(0,255,0),thickness=10)
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(0,0,200),thickness=10)
plt.imshow(blank_img)
```

```python
#draw a circle
cv2.circle(blank_img,center=(100,210), radius=50,color=(255,0,0),thickness=10)
plt.imshow(blank_img)
```

```python
#draw a circle and fill it with color
cv2.circle(blank_img,center=(400,40), radius=50,color=(255,0,0),thickness=-1)
plt.imshow(blank_img)
```

```python
#draw lines
cv2.line(blank_img,pt1=(0,0),pt2=(512,512),color=(0,255,0),thickness=10)
plt.imshow(blank_img)
```

```python
#writting on a shape
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img,'Hello',org=(10,500),fontFace=font,fontScale=4,color=(255,255,255),thickness=3, lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

```python
#drawing custom polygons
blank_img = np.zeros(shape=(512,512,3), dtype=np.int32)
plt.imshow(blank_img)
```

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
```

```python
vertices.shape
```

```python
vertices
```

```python
#convert vertices to shape 3
pts = vertices.reshape(-1,1,2)
```

```python
pts.shape
```

```python
pts
```

```python
cv2.polylines(blank_img,[pts],isClosed=True,color=(255,255,255),thickness=5)
plt.imshow(blank_img)
```
