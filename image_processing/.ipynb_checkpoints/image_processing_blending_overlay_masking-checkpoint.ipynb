```python
import cv2

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path1 = '/Users/robertkigobe/Documents/My_Research/My_Python/Computer_vision/images/norumbega.png'
path2 = '/Users/robertkigobe/Documents/My_Research/My_Python/Computer_vision/images/new_norumbega.jpg'
img1 = cv2.imread(path1)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread(path2)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
#resize images for calculations
img1 = cv2.resize(img1,(1500,1500))
img2 = cv2.resize(img2,(1500,1500))
```

```python
#blend images addweighted works only if images are the same size
blended = cv2.addWeighted(src1=img1,alpha=0.8,src2=img2,beta=0.5,gamma=0)
```

```python
plt.imshow(blended)
```

```python
#Overlay
img1 = cv2.imread(path1)
larger_img = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread(path2)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
smaller_img = cv2.resize(img2,(1200,1200))
```

```python
plt.imshow(larger_img)
```

```python
plt.imshow(smaller_img)
```

```python
x_offset = 0
y_offset = 0
```

```python
x_end = x_offset + smaller_img.shape[1]
y_end = y_offset + smaller_img.shape[0]
```

```python
larger_img[y_offset:y_end,x_offset:x_end] = smaller_img
```

```python
plt.imshow(larger_img)
```
