```python
#do a good size plot
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
#read in an image in grayscale
img = cv2.imread(path2,0)
show_pic(img)
```

```python
#do some thresholding

#choose the halfway point 127.5 meaning any value below 127 is 0 else 1
ret,thresh1=cv2.threshold(img,180,255,cv2.THRESH_BINARY)
show_pic(img)
```

```python
#use adaptive threshold
ths2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8)
show_pic(img)
```

```python
#blend the thresholds
blended = cv2.addWeighted(src1=thresh1,alpha=0.6, src2=ths2,beta=0.4, gamma=0)
show_pic(blended)
```
